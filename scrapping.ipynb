## Scrapping

```python
def go_to_url(url):   
    #an exception might be thrown, so the code should be in a try-except block
    try:
        #use the browser to get the url. This is suspicious command that might blow up.
        page=requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})                             # this might throw an exception if something goes wrong.
        return page

    except Exception as e:                                   # this describes what to do if an exception is thrown
        error_type, error_obj, error_info = sys.exc_info()      # get the exception information
        print ('ERROR FOR LINK:',url)                          #print the link that cause the problem
        print (error_type, 'Line:', error_info.tb_lineno)     #print error info and line that threw the exception
                                                 #ignore this page. Abandon this and go back.
    time.sleep(2) 
```

fonction permettant de faire scroller automatiquement une page web

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from random import *

#launcher -> ouvre une nouvelle page web avec l'url indiquée
driver = webdriver.Firefox(executable_path= r"C:\Users\utilisateur\Downloads\geckodriver.exe")

driver.get('https://www.lejdd.fr/election-presidentielle')

time.sleep(5)

# enlève le pop-up des cookies
stop_cookies = driver.find_element_by_id("didomi-notice-agree-button")
stop_cookies.click()

time.sleep(3)

# focus sur le corps de la page web
element = driver.find_element_by_tag_name('body')

# automatisation du scrolling avec aléatoire pour duper les bots
timeout = 60 * 30

timeout_start = time.time()

while time.time() < timeout_start + timeout:
    element.send_keys(Keys.PAGE_DOWN)
    time.sleep(randint(1,6))
```

```python
# récupération des urls dans la page déroulée
elems = driver.find_elements_by_xpath("//h3[contains(@class, 'titre')]/a[@href]")
urls = []
#stockage des urls dans une liste
for elem in elems:
    urls.append(elem.get_attribute("href"))
```

```python
upperframe=[]  

frame=[]
links = urls #soup.find_all('h3',attrs={'class':'titre'})
    #print(len(links))
filename = "leJDD.csv"
f = open(filename,"w", encoding = 'utf-8')
headers = "Journal,Lien,Date,Source,Contenu\n"
f.write(headers)
compteur = 0    
for j in links:
    compteur += 1
    print('lien n°', compteur,' :', j)
    Journal = 'LeJDD'
    #Lien = "https://www.lejdd.fr"
    #Lien += j.find('a')['href'].strip('\n').strip()#"div",attrs={'class':'m-statement__quote'}
    Lien = j

    article = go_to_url(Lien)
    art_soup = BeautifulSoup(article.text,'html.parser')

    Date = art_soup.find('time').get_text(separator=', ', strip=True)#.find('footer').text[-14:-1].strip()#'div',attrs={'class':'m-statement__body'}
    print(art_soup.find('time').getText(separator=' ', strip=True))
    print(Date)
        
        #Source = auteur de l'article
    if art_soup.find('li',attrs={'class':'auteur'}) != None:
        print('aut1:', art_soup.find('li',attrs={'class':'auteur'}))
        Source = art_soup.find('li',attrs={'class':'auteur'}).find('a').text.strip('\n').strip()
    elif art_soup.find('li',attrs={'class':'no-auteur'}) != None:
            #print('aut2:', art_soup.find('li',attrs={'class':'no-auteur'}))
        Source = art_soup.find('li',attrs={'class':'no-auteur'}).text.strip('\n').strip()
    else:
        Source = None

    Contenu = art_soup.find('div', attrs ={'class':'chapo'}).find('p').text
        #print(Contenu)
    bis = art_soup.find('div', attrs ={'class':'cnt' , 'data-poool':'80'}).get_text(separator=', ', strip=True)
        #print(bis)
    Contenu += bis
        #print(Contenu)

    frame.append((Journal,Lien,Date,Source,Contenu))
        #f.write(Journal.replace(","," ")+","+Lien+","+Date(","," ")+","+Source.replace(","," ")+","+Contenu.replace(","," ")+"\n") #+","+Label.replace(",","^")
    f.write(Journal+","+Lien+","+Date+","+Source+","+Contenu+"\n")
upperframe.extend(frame)
f.close()
```
